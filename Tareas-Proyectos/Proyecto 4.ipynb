##### Valor total de ventas

Empezaremos con una agrupacion, esta agrupacion se conformara de separar a los manufactureros con el total de valor de ventas(Ganancias)

```python
df_sales_manu = df.groupby('MANUFACTURER')['TOTAL_VALUE_SALES'].sum().round(2)
df_sales_manu
```

Como podemos ver en la tabla, a simple vista ya nos podemos hacer un pequeño adelanto de cuales son tienen mas cuota de mercado o ganacias

```python
# Grafico de pie
series = df_sales_manu.sort_values(ascending=False)
labels = series.index.tolist()
values = series.values

plt.figure(figsize=(7, 5))
plt.pie( values, labels=labels, autopct='%1.1f%%', startangle=45, colors=sns.color_palette("tab10", n_colors=len(values)), wedgeprops={"edgecolor": "white"})
plt.title("Distribución de ventas por Manufacturer")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

Insigth: El manufacturer con mas porcentaje de ventas en el mercado es Reckit

##### Total de unidades vendidas

De igual forma que el valor de ventas, obtendremos el total de las unidades vendidas por cada manufacturardor

```python
df_unit_sales_manu = df.groupby('MANUFACTURER')['TOTAL_UNIT_SALES'].sum().round(2)
df_unit_sales_manu.sort_values(ascending=False)
```

```python
# Grafico de barras
series = df_unit_sales_manu.sort_values(ascending=False)
labels = series.index.tolist()
values = series.values

plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=sns.color_palette("tab10", n_colors=len(values)))
plt.title("Distribución de Unidades Vendidas por Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Total Unidades Vendidas")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Se realizo una pequeña comparacion de las unidades vendidas.

Lo que observamos, es una significatica distancia de ventas entre P&G a Jabonera, y de Jabonera a Reckitt

Volviendo al valor de ventas, realizaremos un desglose anual para ver de mejor manera el desarrollo de cada manufacturador

```python
# Tabla pivot del total de ventas por año
pvt_df_sales_year = pd.pivot_table(df, values='TOTAL_VALUE_SALES', index='MANUFACTURER', columns='YEAR', aggfunc='sum', fill_value=0).round(2)
pvt_df_sales_year
```

```python
# Análisis de productos más vendidos por región
pvt_products_region = pd.pivot_table(df, values='TOTAL_VALUE_SALES', index='REGION', columns='BRAND', aggfunc='sum', fill_value=0).round(2)

# Obtener top 5 productos (marcas) por región
top_products_region = pvt_products_region.apply(lambda x: x.nlargest(5).index.tolist(), axis=1)

# Visualizar los top 10 productos por región
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for idx, region in enumerate(pvt_products_region.index):
    top_brands = pvt_products_region.loc[region].nlargest(10)
    axes[idx].barh(top_brands.index, top_brands.values, color=sns.color_palette("viridis", len(top_brands)))
    axes[idx].set_title(f'Top 10 Marcas - {region}')
    axes[idx].set_xlabel('Total Valor de Venta')
    axes[idx].invert_yaxis()

plt.tight_layout()
plt.show()
```

A lo obtenido, se ven 3 marcas top, de las cuales Vanish tiene una cuota de mercado regional muy marcada, ya que en todas las regiones, este domina sin| comparacion

```python
# Tabla resumen de top 3 productos por región
print("Top 3 Marcas por Región:")
for region in pvt_products_region.index:
    top_3 = pvt_products_region.loc[region].nlargest(3)
    print(f"\n{region}:")
    for brand, value in top_3.items():
        print(f"  {brand}: {value:.2f}")
```
