## Matriz correlacional / HeatMap

Pregunta a resolver= 
- ¿Qué tan fuerte es la relación entre el valor de mercado (Value) y el rendimiento (Overall)?
- ¿La edad (Age) afecta el potencial (Potential) o el salario (Wage)?
- ¿Los atributos físicos (Height, Weight) tienen alguna correlación con el rendimiento o el valor

### ¿Qué tan fuerte es la relación entre el valor de mercado (Value) y el rendimiento (Overall)?

```python
df['relacion_valor_rendimiento'] = df['Value'] / df['Overall']
```

```python
# Seleccionamos las columnas numericas, excluyendo las no relevantes para el analisis
df_num = df.select_dtypes(include=['number']).drop(columns=['ID', 'Age', 'Potential', 'Height', 'Weight', 'Joined','Skill Moves', 'International Reputation'])
```

```python
df_num.corr()
```

```python
fig = px.imshow(df_num.corr(), text_auto=True)
fig.show()
```

### ¿La edad (Age) afecta el potencial (Potential) o el salario (Wage)?

```python
df_potencial = df[['Age', 'Potential', 'Wage']]
df_potencial.corr()
```

```python
px.imshow(df_potencial.corr(), text_auto=True).show()
```

## Grafico Correlacional
Cuál es la relación entre la edad y el overall?

```python
df_corr = df
df_corr
```

```python
fig = px.scatter(df_corr, x='Age', y='Overall', color='Potential', title='Relacion entre Edad y Overall').show()
```

A lo que spuede observar, es que los jugadores mas jovenes tienen 'Overall' mas variados, a diferencias de jugadores 30+, que cuentan con un una puntacion mas alta y estable.

De igual forma, se puede apreciar una tendencia negativa algo debil mientras mas años tengan los jugadores

En conclusion la relacion no es lineal ni fuerte

## Grafico de Barras x Club

```python
df_bar = df
```

```python
df_bar = df_bar['Club'].value_counts().rename('Numero').reset_index()
df_bar
```

```python
fig = px.bar(df_bar, x='Club', y='Numero', title='Numero de Jugadores por Club')
fig.show()
```
