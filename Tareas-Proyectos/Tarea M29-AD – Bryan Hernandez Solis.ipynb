## Uso de librerias

```python
# Realizar web scraping en la p√°gina de Despegar
despegar_url = "https://www.us.despegar.com"
response = requests.get(despegar_url)
soup = BeautifulSoup(response.text, "html.parser")

# Extraer t√≠tulos de las ofertas principales (por ejemplo, t√≠tulos de viajes)
offers = soup.find_all('h2')
for offer in offers:
    print(offer.get_text(strip=True))
```

```python


# URL del sitio a scrapear
url = "https://www.us.despegar.com/"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    # 1. Realizar la petici√≥n GET para obtener el HTML de la p√°gina
    response = requests.get(url, headers=headers)
    
    # Lanza un error si la petici√≥n no fue exitosa (ej. error 404 o 500)
    response.raise_for_status()

    # 2. Parsear el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # 3. Encontrar los elementos que contienen la informaci√≥n
    # Tras inspeccionar la p√°gina, los t√≠tulos de las ofertas est√°n en etiquetas <h3>
    # dentro de contenedores con la clase 'offer-card-title'.
    # Nota: Las clases pueden cambiar, esto funcionaba al momento de escribir el c√≥digo.
    offer_titles = soup.find_all('h3', class_='offer-card-title')

    print("--- T√≠tulos de Ofertas Encontrados ---\n")

    if offer_titles:
        # 4. Iterar sobre los resultados y mostrar la informaci√≥n
        for index, title_tag in enumerate(offer_titles):
            # .get_text(strip=True) limpia el texto de espacios extra
            title_text = title_tag.get_text(strip=True)
            print(f"{index + 1}. {title_text}")
    else:
        print("No se encontraron t√≠tulos de ofertas con la clase 'offer-card-title'.")
        print("Es posible que la estructura del sitio web haya cambiado.")

except requests.exceptions.RequestException as e:
    print(f"Error al intentar conectar con el sitio web: {e}")
except Exception as e:
    print(f"Ocurri√≥ un error inesperado: {e}")

```
