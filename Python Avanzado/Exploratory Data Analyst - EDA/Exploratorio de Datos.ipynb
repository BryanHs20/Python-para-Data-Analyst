## Analisis Correlacional
(Analisis del grado de interelacional lineal de una variable con otra)

```python
df.corr('pearson')
```

```python
plt.figure(figsize=(14,10))
# Selecciona unicamente las coulmnas numericas por correlacion
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_df = numeric_df.corr(method='pearson')

df_lt = corr_df.where(np.tril(np.ones(corr_df.shape)).astype(bool))
hmap = sns.heatmap(df_lt, cmap='YlGnBu', annot=True)
```

```python
print(df.columns)
```

```python
cols_num = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']
```

```python
# Grafico exploratorio de todas las columnas
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(18, 16))
for i, column in enumerate(cols_num):
    sns.distplot(df[column], ax=axes[i//4, i%4], kde=False)
```

## Agrupamiento

```python
# Agrupamiento general por variables
df.groupby('waterfront')[['price', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built', 'grade']].median()
```

```python
df.groupby('yr_built')[['price', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'grade']].median()
```

```python
df.groupby(['floors'])[['price', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'grade']].median()['price'].plot.bar()
```

## Analisis Bivariado

```python
pd.crosstab(df['waterfront'], df['floors'])
```

```python
pd.crosstab(df['grade'], df['floors'], margins=True, margins_name='Total')
```

```python
dfc1 = pd.crosstab(df['grade'], df['floors'], values=df.price, aggfunc='mean', margins=True, margins_name='Total')
dfc1 = dfc1.replace(np.nan, 0)
dfc1
```

```python
dfc1.style.background_gradient(cmap='Greens')\
    .set_properties(**{'font-size': '10px'})
```
