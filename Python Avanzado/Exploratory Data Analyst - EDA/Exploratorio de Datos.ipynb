```python
df1 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_January_2019.csv')
df2 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_February_2019.csv')
df3 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_March_2019.csv')
df4 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_April_2019.csv')
df5 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_May_2019.csv')
df6 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_June_2019.csv')
df7 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_July_2019.csv')
df8 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_August_2019.csv')
df9 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_September_2019.csv')
df10 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_October_2019.csv')
df11 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_November_2019.csv')
df12 = pd.read_csv('../../../Archivos-Analisis/Modulo 33- EDA/sales_month/Sales_December_2019.csv')
```

```python
df1
```

```python
# Genera un dataset final donde se van a compilar todos los anteriores
df = df1

# Genera un campo de fecha
df['Month'] = '2019-01-01'
```

```python
df
```

```python
# Concatenamos los demas meses
# Feb
df = pd.concat([df,df2], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-02-01')

# Mar
df = pd.concat([df,df3], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-03-01')

# Abr
df = pd.concat([df,df4], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-04-01')

# May
df = pd.concat([df,df5], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-05-01')

# Jun
df = pd.concat([df,df6], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-06-01')

# Jul
df = pd.concat([df,df7], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-07-01')

# Agos
df = pd.concat([df,df8], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-08-01')

# Sep
df = pd.concat([df,df9], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-09-01')

# Oct
df = pd.concat([df,df10], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-10-01')

# Nov
df = pd.concat([df,df11], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-11-01')

# Dic
df = pd.concat([df,df12], ignore_index=True)
df['Month'] = df['Month'].fillna('2019-12-01')
```

## Masajeo de Datos

```python
# Eliminacion de nulos
df = df.dropna(how='all')
```

```python
df['Order Date']
```

```python
# Eliminamos las cadenas especiales para la fecha de pedido (Renglon anterior)
df = df[df['Order Date'].str[0:2] != 'Or']
```

```python
# Cambiamos los tipos de datos, debido a que la mayoria son del tipo Object, y eso nos puede dificultad las cosas
df['Month'] = pd.to_datetime(df["Month"])
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
# Se tienen que tner cuidado con las cantidades
df['Quantity Ordered'].unique()
```

```python
# Conversion de las columnas numericas
df['Order ID'] = df['Order ID'].astype('string')

df['Product'] = df['Product'].astype('string')

df['Quantity Ordered'] = df['Quantity Ordered'].fillna(0)
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int64')

df['Price Each'] = df['Price Each'].astype('float')

df['Purchase Address'] = df['Purchase Address'].astype('string')
```

```python
print(df.dtypes)
```

```python
# Numeros de datos vacios por columna
missing_data = df.isnull().sum()
missing_data
```

```python
# Calcula el % de datos no existentes
# Cuantos datos tenemos y cuantos faltan
total_cells = np.prod(df.shape)
total_missing = missing_data.sum()

# Calculamos el procentaje
percent_missing = (total_missing / total_cells) * 100
print(f'Porcentaje de Datos quie nmo existen: {percent_missing:.2f}%')
```

```python
# Eliminaremos los datos donde la cantidad sea cero
df[df['Quantity Ordered'] == 0]
```

```python
# Actualizamos el DF, dandole como parametro que se actualice el campo 'Quantity Ordered' siempre y cuando sea > 0
df = df[df['Quantity Ordered'] > 0]
```

```python
# Corroboramos los datos faltantes nuevamente
missing_data = df.isnull().sum()
missing_data
```

```python
# Generaremos la columna de 'Zip code' a partir de la culumna Purchase Address
df['Purchase Address']
```

```python
# Se genera la columna tomando los ultimos 5 digitos de la direccion
df['Zip Code'] = df['Purchase Address'].str[-6:]
```

```python
# Total de zip codes analisados
df['Zip Code'].unique()
```

```python
# Lo cambiamos al tipo de dato string
df['Zip Code'] = df['Zip Code'].astype('string')
```

```python
df.info()
```

```python
# Encotramos el numero de zipcodes
num_zipcode = df['Zip Code'].unique().__len__()
num_zipcode
```

## Descubrimientos (Insigths)

```python
# Generamos un campo de ventas ya que no existe
df['Total Sales'] = df['Quantity Ordered'] * df['Price Each']
df.sample(5)
```

- Top productos en ventas
- Productos con mayor crecimiento
- Vistas de ventas de productos por cada uno de ellos

```python
#Productos con mayor y menor crecimiento
# Obtenemos primero las ventas por producto
pd.crosstab(df['Product'], df['Month'], values=df['Total Sales'], aggfunc='sum', margins=True).round(2)
```

```python
pd.crosstab(df['Group'], df['Month'], values=df['Total Sales'], aggfunc='sum', margins=True).round(2)
```

```python
# Validamos la venta de enero de un monitor
df[df['Month'] == '2019-01-01']['Total Sales'].sum()
# Se compara el total de enero con el total por filtro
```

```python
# Venta de enero y Monitor especifico
df[(df['Month'] == '2019-01-01') & (df['Product'] == '20in Monitor')]['Total Sales'].sum()
# el resultado sera igual al global
```

```python
pd.pivot_table(df, values='Total Sales', index=['Month'], columns=['Group'], aggfunc='sum').fillna(0).round(2).reset_index()
```

```python
dfsales = pd.pivot_table(df, values='Total Sales', index=['Month'], columns=['Product'], aggfunc='sum').fillna(0).round(2).reset_index()
```

```python
# Cuadro con el porcentaje de ventas de cada producto para cada mes
dfsales[list(dfsales.columns)[1:20]].pct_change().round(2).fillna(0)
```

```python
# Definimos a los productos de mas venta como los que han tenido mayor crecimiento promedio en el ultimo año
# Definimos
df_sales_pct = dfsales[list(dfsales.columns)[1:20]].pct_change().round(2).fillna(0)
```

```python
# Obtenemos los productos con amyor crecimiento del ultimo a;o
df_sales_pct.mean().sort_values(ascending=False).round(2)
```

```python
df_sales_pct.plot()
```
